Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
Answer 4-
In Pandas, a DataFrame is a 2-dimensional labeled data structure that consists of rows and columns, similar to a spreadsheet or a SQL table. It is a widely used data structure in data analysis and manipulation tasks in Python. Each column in a DataFrame can contain different data types such as integers, floats, strings, and others.

On the other hand, a Pandas Series is a one-dimensional labeled array that can hold data of any type, similar to a column in a DataFrame. A Series can be created from a list, tuple, dictionary, or any other array-like object.

The key difference between a DataFrame and a Series is that a DataFrame is a 2-dimensional data structure that can hold multiple columns, while a Series is a 1-dimensional data structure that can hold only one column.

Q7. Create a DataFrame using multiple Series. Explain with an example.
Answer 7:-
You can create a DataFrame from multiple Series objects by adding each series as a columns.

By using concat() method you can merge multiple series together into DataFrame. This takes several params, for our scenario we use list that takes series to combine and axis=1 to specify merge series as columns instead of rows. Note that using axis=0 appends series to rows instead of columns.

```python
import pandas as pd
courses=pd.Series(["Python","Spark","Hadoop"],name='Courses')
fees=pd.Series([30000,25000,23576],name='Fees')
discount=pd.Series([1000,2000,500],name='Discount')

# combining of multiple series
df1=pd.concat([courses,fees,discount],axis=1)
df1
```
