Compared to nominal data, ordinal data have some kind of order that is not present in nominal data.

There are 4 levels of measurement, which can be ranked from low to high: Nominal: the data can only be categorized. Ordinal: the data can be categorized and ranked. Interval: the data can be categorized and ranked, and evenly spaced. Ratio: the data can be categorized, ranked, evenly spaced and has a natural zero.

Nominal level	
You can categorize your data by labelling them in mutually exclusive groups, but there is no order between the categories.
Examples of nominal scales	
City of birth
Gender

Ans 6:-importance:
Knowing the level of measurement of your variables is important for two reasons.  Each of the levels of measurement provides a different level of detail.  Nominal provides the least amount of detail, ordinal provides the next highest amount of detail, and interval and ratio provide the most amount of detail.
The second reason levels of measurement are important to know is because different statistical tests are appropriate for variables with different levels of measurement.  For example, chi-square tests of independence are most appropriate for nominal level data.  The Mann-Whitney U test is most appropriate for an ordinal level dependent variable and a nominal level independent variable.  An ANOVA is most appropriate for a continuous level dependent variable and a nominal level independent variable.

In statistics, measures of central tendency and variability are used to describe the distribution of a dataset. Measures of central tendency indicate the most typical or representative value of a dataset, while measures of variability indicate the degree to which the values in the dataset vary from one another. Some common measures of central tendency and variability include:

Mean: The mean is calculated by adding up all of the values in the dataset and dividing by the number of values. It represents the average value of the dataset. The mean is a useful measure of central tendency because it takes into account all of the values in the dataset and can be used to compare different datasets. However, the mean can be influenced by extreme values or outliers, which can skew the result.

Median: The median is the middle value in a dataset when the values are arranged in order from smallest to largest. It represents the value that separates the dataset into two equal halves. The median is a useful measure of central tendency because it is not influenced by extreme values or outliers. It can be used to describe datasets that are not normally distributed.

Mode: The mode is the value that appears most frequently in a dataset. It represents the most common value in the dataset. The mode is a useful measure of central tendency for datasets with a large number of repeated values. However, it may not be useful for datasets where there is no clear mode or where there are several values that appear with similar frequency.

Range: The range is the difference between the largest and smallest values in a dataset. It represents the spread of the dataset. The range is a useful measure of variability because it gives an idea of how much the values in the dataset vary from one another. However, it is influenced by extreme values or outliers.

Variance: Variance is a measure of how spread out the values in a dataset are from the mean. It is calculated by taking the sum of the squared differences between each value and the mean, and dividing by the number of values. Variance is a useful measure of variability because it takes into account all of the values in the dataset and can be used to compare different datasets. However, it is influenced by extreme values or outliers.

Standard Deviation: The standard deviation is the square root of the variance. It represents the average distance that the values in the dataset are from the mean. The standard deviation is a useful measure of variability because it takes into account all of the values in the dataset and can be used to compare different datasets. It is also useful for describing datasets that are normally distributed. However, it is influenced by extreme values or outliers.
